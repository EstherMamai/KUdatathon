```python
Weathers=df['Weather'].value_counts().reset_index()
Weathers.columns=['Weather','Count']
Weathers
```

```python
sns.set(rc={'figure.figsize':(10,8)})
plt.xticks(rotation=90)
sns.lineplot(x=Weathers['Weather'],y=Weathers['Count'],data=Weathers)
```

```python
plt.figure(figsize=(12, 6)) 
plt.xticks(rotation=90)  
plt.title('Weather vs Relative Humidity')
sns.barplot(x=df['Weather'], y=df['Rel Hum_%'])
plt.xlabel('Weather Condition')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.xticks(rotation=90)  
plt.title('Weather vs Relative Humidity')
sns.barplot(x=df['Weather'], y=df['Wind Speed_km/h'])
plt.xlabel('Weather Condition')
plt.ylabel('Wind Speed')
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.xticks(rotation=90)  
plt.title('Weather vs Visibility')
sns.barplot(x=df['Weather'], y=df['Visibility_km'])
plt.xlabel('Weather Condition')
plt.ylabel('Visibility')
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.xticks(rotation=90)  
plt.title('Weather vs Temperature ')
sns.barplot(x=df['Weather'], y=df['Temp_C'])
plt.xlabel('Weather Condition')
plt.ylabel('Temperature')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,8)})
plt.xticks(rotation=90)
sns.lineplot(x=df['Weather'],y=df['Press_kPa'],data=df)
```

```python
df = df.drop(['Weather'], axis = 1)
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
#assign correlation matrix
relation=df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(data=relation)
plt.title('Correlation of columns in dataframe')
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
plt.show()
```
