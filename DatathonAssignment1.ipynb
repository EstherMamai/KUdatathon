```python
#Analyzing frequency of purchase
purchase_frequency = df1.groupby('Customer ID')['Order ID'].nunique().reset_index()


purchase_frequency.columns = ['Customer ID', 'Purchase Frequency']


print(purchase_frequency)
```

```python

# Total purchase frequency
total_purchase_frequency = purchase_frequency['Purchase Frequency'].sum()

# Total number of customers
total_customers = purchase_frequency['Customer ID'].nunique()

# Mean frequency of purchase
mean_purchase_frequency = total_purchase_frequency / total_customers

print("Mean Frequency of Purchase:", mean_purchase_frequency)
```

```python
# Identify high-frequency customer
# Since the mean frequency per user is 16, we assumed high frequency purchase was from 20
high_frequency_customers = profit_per_customer[profit_per_customer['Purchase Frequency'] > 25]

total_revenue = df1['Sales'].sum() #/ df1['Customer ID'].nunique()
total_profitability = df1['Profit'].sum() #/ df1['Customer ID'].nunique()

# Compare metrics
high_freq_revenue = high_frequency_customers['Sales'].sum()
high_freq_profitability = high_frequency_customers['Profit'].sum()

revenue=high_freq_revenue/total_revenue
profitability=high_freq_profitability/total_profitability

print("Total Revenue per Customer:", total_revenue)
print("Total Profitability per Customer:", total_profitability)
print("Total Revenue from High-Frequency Customers:", high_freq_revenue)
print("Total Profitability from High-Frequency Customers:", high_freq_profitability)
print("Percentage profitability from high frequency customers:", profitability)
print("Percentage revenue from high frequency customers:",revenue)
```

```python
# Calculate total revenue and profitability per segment
profit_per_segment = df1.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Calculate profit margin (profit as a percentage of sales)
profit_per_segment['Profit Margin'] = (profit_per_segment['Profit'] / profit_per_segment['Sales']) * 100

# Calculate overall revenue and profit margin
total_revenue = df1['Sales'].sum()
total_profit = df1['Profit'].sum()
overall_profit_margin = (total_profit / total_revenue) * 100

# Display the results
print("Overall Revenue:", total_revenue)
print("Overall Profit:", total_profit)
print("Overall Profit Margin:", overall_profit_margin)

print("\nRevenue and Profitability per Segment:")
print(profit_per_segment)

```

```python
plt.figure(figsize=(10, 6))

# Bar chart for Revenue
plt.subplot(1, 2, 1)
plt.bar(profit_per_segment['Segment'],profit_per_segment ['Sales'], color='skyblue')
plt.title('Total Revenue by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Total Revenue')

# Bar chart for Profit Margin
plt.subplot(1, 2, 2)
plt.bar(profit_per_segment ['Segment'], profit_per_segment['Profit Margin'], color='lightgreen')
plt.title('Profit Margin by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Profit Margin')

plt.tight_layout()
plt.show()
```

```python

# Calculate total sales per country
total_sales_per_country = df1.groupby('Country')['Sales'].sum().reset_index()

# Find the country with the highest total sales
top_sales_country = total_sales_per_country.loc[total_sales_per_country['Sales'].idxmax()]

print("Top Sales Country:", top_sales_country['Country'])
print("Total Sales:", top_sales_country['Sales'])
```

```python
# Converting 'Order Date' to datetime
df1['Order Date'] = pd.to_datetime(df1['Order Date'])

# Extract the year from 'Order Date'
df1['Year'] = df1['Order Date'].dt.year

# Group the data by year and product category, and calculate total profit
profit_per_year_category = df1.groupby(['Year', 'Category'])['Profit'].sum().reset_index()

# Sort the data by year and profit in descending order
profit_per_year_category = profit_per_year_category.sort_values(by=['Year', 'Profit'], ascending=[True, False])

# Get the top 5 profit-making product types for each year
top_5_profit_per_year = profit_per_year_category.groupby('Year').head(5)

print(top_5_profit_per_year)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

sns.barplot(x='Category', y='Profit', hue='Year', data=top_5_profit_per_year)

plt.title('Top 5 Profit-Making Product Categories for Each Year')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')

plt.xticks(rotation=45)

plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
# Calculate total profit for each product type within each year
profit_by_product_yearly = df1.groupby(['Year', 'Product Name'])['Profit'].sum().reset_index()
top_profit_products_by_year = profit_by_product_yearly.groupby('Year').apply(lambda x: x.nlargest(5, 'Profit')).reset_index(drop=True)
```

```python

plt.figure(figsize=(12, 6))
for year, data in top_profit_products_by_year.groupby('Year'):
    plt.bar(data['Product Name'], data['Profit'], label=year)

plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.title('Top 5 Profit-Making Product Types on a Yearly Basis')
plt.legend(title='Year')
plt.show()
```

```python
# Mapping dictionary for subcategories and their corresponding SDGs
subcategory_sdg_mapping = {
    'Accessories': 'SDG 12: Responsible Consumption and Production',
    'Chairs': 'SDG 8: Decent Work and Economic Growth',
    'Phones': 'SDG 9: Industry, Innovation, and Infrastructure',
    'Copiers': 'SDG 9: Industry, Innovation, and Infrastructure',
    'Tables': 'SDG 12: Responsible Consumption and Production',
    'Binders': 'SDG 12: Responsible Consumption and Production',
    'Supplies': 'SDG 12: Responsible Consumption and Production',
    'Appliances': 'SDG 7: Affordable and Clean Energy',
    'Machines': 'SDG 9: Industry, Innovation, and Infrastructure',
    'Bookcases': 'SDG 4: Quality Education',
    'Storage': 'SDG 9: Industry, Innovation, and Infrastructure',
    'Furnishings': 'SDG 12: Responsible Consumption and Production',
    'Art': 'SDG 12: Responsible Consumption and Production',
    'Paper': 'SDG 12: Responsible Consumption and Production',
    'Envelopes': 'SDG 12: Responsible Consumption and Production',
    'Fasteners': 'SDG 12: Responsible Consumption and Production',
    'Labels': 'SDG 12: Responsible Consumption and Production'
}

# Classify subcategories into SDGs
subcategories_with_sdgs = {subcategory: subcategory_sdg_mapping[subcategory] for subcategory in unique_subcategories}

# Print the result
for subcategory, sdg in subcategories_with_sdgs.items():
    print(f"{subcategory}: {sdg}")
```

```python
sdg_mapping = {
    'Appliances': 'Poverty Reduction',
    'Machines': 'Poverty Reduction',
    'Paper': 'Poverty Reduction',
    'Chairs': 'Gender Equality',
    'Binders': 'Gender Equality',
    'Envelopes': 'Gender Equality',
    'Accessories': 'Environmental Sustainability',
    'Phones': 'Environmental Sustainability',
    'Copiers': 'Environmental Sustainability',
    'Tables': 'Environmental Sustainability',
    'Storage': 'Environmental Sustainability',
    'Furnishings': 'Environmental Sustainability',
    'Art': 'Environmental Sustainability',
    'Fasteners': 'Environmental Sustainability',
    'Labels': 'Environmental Sustainability'
}

# Classify subcategories into SDGs
subcategories_with_sdgs = {}
for subcategory in unique_subcategories:
    sdg = sdg_mapping.get(subcategory, 'Other')
    if sdg in subcategories_with_sdgs:
        subcategories_with_sdgs[sdg].append(subcategory)
    else:
        subcategories_with_sdgs[sdg] = [subcategory]

# Print the classification
for sdg, subcategories in subcategories_with_sdgs.items():
    print("SDG:", sdg)
    print("Subcategories:", subcategories)
    print()
```
